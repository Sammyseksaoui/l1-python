```python
from bs4 import BeautifulSoup
import pandas as pd

import requests

response = requests.get("https://www.auchan.fr/joyeuses-fetes/idees-cadeaux/ca-13224333")
soup = BeautifulSoup(response.text,features="html.parser")
prod = soup.find_all ("p",{'class':"product-thumbnail__description"})
print (prod)
produits=[]
for e in prod :
    e=e.text
    e=e.replace('\n','')
    produits.append(e)
    
print (produits)

prix = soup.find_all("div",{'class' :"product-price bolder red-txt"})
print (prix)

prix_initials = []
for e in prix :
    e = e.text

    prix_initials.append (e)
print (prix_initials)
```
